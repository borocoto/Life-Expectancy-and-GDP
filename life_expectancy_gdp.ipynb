```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
print(df.head())
print(df.info())
print(df['Country'].unique())
```

```python
countries = df['Country'].unique()
mean_expectancies = []
for country in countries:
    mean_exp = np.mean(df['Life expectancy at birth (years)'][df['Country'] == country])
    mean_expectancies.append(mean_exp)
    
mean_gdps = []
for country in countries:
    mean_gdp = np.mean(df['GDP'][df['Country'] == country])
    mean_gdps.append(mean_gdp)
    
print(countries)
print(mean_gdps)
print(mean_expectancies)
```

```python
for i in range(len(countries)):
    plt.scatter(mean_expectancies[i], mean_gdps[i], s=80)
    plt.legend(countries)
plt.ylabel('GDP')
plt.xlabel('Life Expectancy')
plt.title('Mean Life Expectancy and GDP per Country')
plt.show()
plt.clf()
```

```python
print("On the chart above mean life expectancy and mean GDP per country are shown. \nAccording to the chart, there is no visible correlation between these two variables in the scale of several countries.")
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.barh(countries, mean_gdps)
plt.xlabel('Mean GDP', fontsize=12)

plt.subplot(2, 1, 2)
plt.barh(countries, mean_expectancies)
plt.xlim(45, 85)
plt.xlabel('Mean Life Expectancy', fontsize=12)

plt.subplots_adjust(hspace=0.5)
plt.show()
plt.clf()
```

```python
print("GDPs in different countries diverse significantly, with the USA having the largest GDP and Zimbabwe and Chile having the least one.")
print("Mean life expectancy in distinct countries is more homogeneous, with only Zimbabwe having visibly less value of this variable.")
```

```python
print("Minimum GDP Value: ", np.min(df['GDP']))
print("Maximum GDP Value: ", np.max(df['GDP']))
sns.histplot(data=df['GDP'], bins=30)
plt.show()
plt.clf()

print("\n")
print("Minimum Life Expectancy Age: ", np.min(df['Life expectancy at birth (years)']))
print("Maximum Life Expectancy Age: ", np.max(df['Life expectancy at birth (years)']))
sns.histplot(data=df['Life expectancy at birth (years)'], bins=30)
plt.show()
plt.clf()
```

```python
print("GDPs in the six countries are distributed between 4.4*10^9 and 1.8*10^13. Most values are placed between 4.4*10^9 and 4000*10^9.")
print("Life expectancies in the six countries are distributed between 44.3 and 81.0. Most values are placed between 75 and 80.")
```

```python
colours = ['#0040ff', '#80ff00', '#ff0040', '#ff8000', '#bfff00', '#00bfff']
plt.figure(figsize=(12, 12))
for i in range(len(countries)):
    plt.subplot(3, 2, (i + 1))
    sns.scatterplot(df['Life expectancy at birth (years)'][df['Country'] == countries[i]], df['GDP'][df['Country'] == countries[i]], color=colours[i])
    plt.title(countries[i])
    plt.subplots_adjust(hspace=0.6)

plt.show()
plt.clf()
```

```python
print("Due to the charts above, there is a strong positive correlation between GDP and life expectancy in each particular country. Also both GDP and life expectancy grow in the course of time in all the six countries.")
```
